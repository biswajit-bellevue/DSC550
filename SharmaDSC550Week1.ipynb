Try to calculate the median user score of all the video games. You will likely run into an error because some of the user score entries are a non-numerical string that cannot be converted to a float. Find and replace this string with NaN and then calculate the median. Then, replace all NaN entries in the user score column with the median value.

```python
# median user score without replacing strings
df["User_Score"].median()
```

```python
# this function converts input value to a float. If the input value is non-numeric data then returns NaN
def convert_to_float(data):
    try:
        return float(data)
    except ValueError:
        return np.nan
```

```python
# replace string values in User_Score column with NaN
df["User_Score"] = df["User_Score"].apply(convert_to_float)
```

```python
# get the median scores
median_score = df["User_Score"].median()
median_score
```

Now, we do not notice any error as the non-numerical string values are replaced with NaN

```python
# replace NaN with median score value
df["User_Score"] = df["User_Score"].replace(np.nan, median_score)
```

```python
# check there are no NaNs
df["User_Score"].isna().sum()
```
